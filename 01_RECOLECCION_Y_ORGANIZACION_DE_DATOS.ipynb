```python
# Ruta completa del archivo CSV
file_path = '/content/drive/MyDrive/Notebooks_Clase_2/content/producto_de_datos/internet_fijo.csv'

df = pd.read_csv(file_path)
```

```python
df.head()
```

**VALORES NULOS O FALTANTES**

```python
print(df.duplicated().sum())
```

```python
#Elimina duplicados:

df = df.drop_duplicates()
print(df.duplicated().sum())
```

```python
# Eliminar filas con valores nulos en las columnas específicas
df = df.dropna(subset=['COD_DEPARTAMENTO', 'COD_MUNICIPIO'])

# Verificar si las filas fueron eliminadas
print(df.isnull().sum())
```

**VALORES DUPLICADOS**

```python
# Listado de columnas y datos faltantes en cada una.

for i in df.columns:
    print ("%20s"%i, np.sum(df[i].isna()))
```

```python
# Verificar si existen valores nulos y cuántos
df.isnull().sum()
```

**CORRECION DE NOMBRE DE COLUMNAS O VARIABLES**

```python
# Diccionario para asignar la fecha de fin de trimestre
ftrimestre = {
    1: '-03-31',
    2: '-06-30',
    3: '-09-30',
    4: '-12-31'
}

# Columna nueva de Fecha Unifica el ultimo dia de cada trimestre
df['FECHA'] = df['AÑO'].astype(str) + df['TRIMESTRE'].map(ftrimestre)

# Fecha a Datatime
df['FECHA'] = pd.to_datetime(df['FECHA'])

# Eliminar las columnas Año y Trimestre
df = df.drop(['AÑO', 'TRIMESTRE'], axis=1)


#Ordena la variable Fecha de primero
cols = ['FECHA'] + [col for col in df.columns if col != 'FECHA']
df = df[cols]

df.columns
```

```python
# Verificar el resultado
print(df.head())
```

```python
df.rename(columns={'No DE ACCESOS': 'SUSCRIPTORES'}, inplace=True)

df.columns
```

```python
#crear un diccionario con la asignación de tecnologías generales
tech_mapping = {
    'FIBER TO THE HOME (FTTH)': 'FIBRA_OPTICA',
    'FIBER TO THE PREMISES': 'FIBRA_OPTICA',
    'FIBER TO THE BUILDING O FIBER TO THE BASEMENT (FTTB)': 'FIBRA_OPTICA',
    'FIBER TO THE NODE (FTTN)': 'FIBRA_OPTICA',
    'FIBER TO THE CABINET (FTTC)': 'FIBRA_OPTICA',
    'OTRAS TECNOLOGÍAS DE FIBRA (ANTES FTTX)': 'FIBRA_OPTICA',
    'CABLE': 'TECNOLOGIA_FIJA',
    'XDSL': 'TECNOLOGIA_FIJA',
    'OTRAS TECNOLOGÍAS INALÁMBRICAS': 'TECNOLOGIA_INALAMBRICA',
    'WIFI': 'TECNOLOGIA_INALAMBRICA',
    'SATELITAL': 'TECNOLOGIA_INALAMBRICA',
    'WIMAX': 'TECNOLOGIA_INALAMBRICA',
    'HYBRID FIBER COAXIAL (HFC)': 'HIBRIDA',
    'OTRAS TECNOLOGÍAS FIJAS': 'OTRAS_TECNOLOGIAS',
    'NA (NO APLICA)': 'OTRAS_TECNOLOGIAS'
}

# Reemplazar las tecnologías con las categorías generales
df['TECNOLOGIA'] = df['TECNOLOGIA'].map(tech_mapping)

# Verificar la nueva categorización
print(df['TECNOLOGIA'].value_counts())
```

```python
# Diccionario de renombrado
renaming_dict = {
    'CORPORATIVO': 'CORPORATIVO',
    'RESIDENCIAL - ESTRATO 1': 'RESIDENCIAL_ESTRATO1',
    'RESIDENCIAL - ESTRATO 2': 'RESIDENCIAL_ESTRATO2',
    'RESIDENCIAL - ESTRATO 3': 'RESIDENCIAL_ESTRATO3',
    'RESIDENCIAL - ESTRATO 4': 'RESIDENCIAL_ESTRATO4',
    'RESIDENCIAL - ESTRATO 5': 'RESIDENCIAL_ESTRATO5',
    'RESIDENCIAL - ESTRATO 6': 'RESIDENCIAL_ESTRATO6',
    'SIN ESTRATIFICAR': 'SIN_ESTRATIFICAR',
    'USO PROPIO INTERNO DEL OPERADOR': 'USO_PROPIO_INTERNO_DEL_OPERADOR'
}

# Reemplazar los valores en la columna 'SEGMENTO'
df['SEGMENTO'] = df['SEGMENTO'].replace(renaming_dict)

# Verificar los cambios
print(df['SEGMENTO'].value_counts())
```

**CORRECCIÓN DE TIPO DE DATOS**

```python
# Verificar los valores únicos de una columna categórica
df['VELOCIDAD_BAJADA'].unique()
df.info()
```

```python
# Reemplazar las comas por puntos
df['VELOCIDAD_BAJADA'] = df['VELOCIDAD_BAJADA'].str.replace(',', '.')

df['VELOCIDAD_SUBIDA'] = df['VELOCIDAD_SUBIDA'].str.replace(',', '.')

# Convertir a tipo numérico (flotante)
df['VELOCIDAD_BAJADA'] = pd.to_numeric(df['VELOCIDAD_BAJADA'], errors='coerce')
df['VELOCIDAD_SUBIDA'] = pd.to_numeric(df['VELOCIDAD_SUBIDA'], errors='coerce')

# Si deseas convertir a enteros, puedes hacerlo después de asegurarte que los datos son válidos
df['VELOCIDAD_BAJADA'] = df['VELOCIDAD_BAJADA'].fillna(0).astype(int)
df['VELOCIDAD_SUBIDA'] = df['VELOCIDAD_SUBIDA'].fillna(0).astype(int)

df.dtypes
```
