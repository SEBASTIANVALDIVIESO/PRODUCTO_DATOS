# **3. LIMPIEZA DE DATOS GENERAL"**



```python
# Paso 1: Eliminar filas con VELOCIDAD_BAJADA = 0 y SUSCRIPTORES > 0
df = df[~((df['VELOCIDAD_BAJADA'] == 0) & (df['SUSCRIPTORES'] > 0))]

# Paso 2: Eliminar filas con VELOCIDAD_SUBIDA = 0 y SUSCRIPTORES > 0
df = df[~((df['VELOCIDAD_SUBIDA'] == 0) & (df['SUSCRIPTORES'] > 0))]

# Paso 3: Si SUSCRIPTORES = 0, igualar VELOCIDAD_BAJADA y VELOCIDAD_SUBIDA a 0
df.loc[df['SUSCRIPTORES'] == 0, ['VELOCIDAD_BAJADA', 'VELOCIDAD_SUBIDA']] = 0

# Verificar el resultado
print("Registros después de la limpieza:")
print(df.head())

# Contar registros restantes
print(f"Cantidad de registros después del proceso: {len(df)}")
```

```python
# Filtrar los registros donde SUSCRIPTORES es 0
suscriptores_cero = df['SUSCRIPTORES'].eq(0).sum()

# Filtrar los registros donde SUSCRIPTORES es 0
suscriptores_cero = df['VELOCIDAD_BAJADA'].eq(0).sum()

# Filtrar los registros donde SUSCRIPTORES es 0
suscriptores_cero = df['VELOCIDAD_SUBIDA'].eq(0).sum()

print(f"Cantidad total de registros con VELOCIDAD_BAJADA es igual a 0: {suscriptores_cero}")
print(f"Cantidad total de registros con SUSCRIPTORES igual a 0: {suscriptores_cero}")
print(f"Cantidad total de registros con VELOCIDAD_SUBIDA es igual a 0: {suscriptores_cero}")

```

```python
# Filtrar registros donde SUSCRIPTORES, VELOCIDAD_BAJADA y VELOCIDAD_SUBIDA son 0
registros_cero = df[(df['SUSCRIPTORES'] == 0) &
                    (df['VELOCIDAD_BAJADA'] == 0) &
                    (df['VELOCIDAD_SUBIDA'] == 0)]

# Contar cuántos registros cumplen esta condición
cantidad_registros_cero = len(registros_cero)

# Mostrar el resultado
print(f"Cantidad de registros con SUSCRIPTORES, VELOCIDAD_BAJADA y VELOCIDAD_SUBIDA igual a 0: {cantidad_registros_cero}")
```

```python
# Convertir la columna 'COD_MUNICIPIO' de float64 a int
df['COD_MUNICIPIO'] = df['COD_MUNICIPIO'].astype('Int64')

# Mostrar los primeros registros para verificar
print(df[['COD_MUNICIPIO']].head())
```

**ELIMINAR COLUMNAS INNCESARIAS**

```python
# ANCHO_BANDA = VELOCIDAD_BAJADA * SUSCRIPTORES
df['ANCHO_BANDA'] = df['VELOCIDAD_BAJADA'] * df['SUSCRIPTORES']

# Verificar los primeros registros para comprobar el resultado
df.head()
```

```python
# COD_MUNICIPIO solo dejar esta variabe, borrar cod depar, depar y muni
#Proveedor, segmento y tecnologia convertir de categoricas a numericas, utilizar Colum transformer

df = df.drop(columns=['COD_DEPARTAMENTO', 'DEPARTAMENTO', 'MUNICIPIO'])

# Verifica el resultado
print(df.head())
```

# **4. IMPORTACION DEL ARCHIVO CSV O/Y EXCEL**

```python
#FRECUENDIAS DE PROVEEDOR

# Ruta completa del archivo CSV
file_path = 'DATOS_LIMPIOS.csv'

df = pd.read_csv(file_path)

# Calcular las frecuencias de PROVEEDOR
frecuencias_proveedor = df["PROVEEDOR"].value_counts(normalize=True).reset_index()
frecuencias_proveedor.columns = ["PROVEEDOR", "FRECUENCIA"]

# Guardar el resultado en un CSV más pequeño
frecuencias_proveedor.to_csv("frecuencias_proveedor.csv", index=False)

print("✅ Archivo 'frecuencias_proveedor.csv' generado correctamente. Sube este archivo para continuar.")
```
